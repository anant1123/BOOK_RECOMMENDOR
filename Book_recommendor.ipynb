```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num-of-rating'},inplace=True)
num_rating
```

```python
avg_rating=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg-rating'},inplace=True)
avg_rating.sample(10)
```

```python
popular=num_rating.merge(avg_rating,on='Book-Title')
popular
```

```python
popular=popular[popular['num-of-rating']>=250].sort_values('avg-rating',ascending=False).head(50)
popular
```

```python
popular=popular.merge(books,on='Book-Title').drop_duplicates('Book-Title').reset_index()[['Book-Title','Book-Author','Image-URL-M','num-of-rating','avg-rating']]
```

```python
padhaku=ratings_with_name.groupby('User-ID').count()['Book-Rating']>100
padhaku=padhaku[padhaku].index
```

```python
kaam_wale=ratings_with_name[ratings_with_name['User-ID'].isin(padhaku)]
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhaku)]
```

```python
famous_books=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=40
```

```python
famous_books=famous_books[famous_books].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]

final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:8]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
